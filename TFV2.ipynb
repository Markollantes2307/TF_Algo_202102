```python
%%writefile ABB.hpp
#include <iostream>
#include <functional>
#include <sstream>
#include <vector>
#include <string>
#include <fstream>
using namespace std;

template <class T>
class Nodo {
public:
	T elemento;
	Nodo* izq;
	Nodo* der;
  Nodo(T elemento){
      this->elemento=elemento;
      izq=der=nullptr;
  }
	Nodo(){  }
};
template <class T>
class ArbolB {
	Nodo<T>* raiz;	
	function<int(T,T)> comparar;
private:
	bool _insertar(Nodo<T>*& nodo, T e) {
		if (nodo == nullptr) {
			nodo = new Nodo<T>(e);
		}
		else if (e < nodo->elemento) {
			return _insertar(nodo->izq, e);
		}
		else if (e >= nodo->elemento) {
			return _insertar(nodo->der, e);
		}
	}
	void _enOrden(Nodo<T>* nodo) {
		if (nodo == nullptr) return;
		_enOrden(nodo->izq);
		cout<<nodo->elemento<<" ";
		_enOrden(nodo->der);
	}
	void _enPreOrden(Nodo<T>* nodo) {
		if (nodo == nullptr) {
			return;
		}
		cout<<nodo->elemento<<" ";
		_enPreOrden(nodo->izq);
		_enPreOrden(nodo->der);
	}
	void _enPosOrden(Nodo<T>* nodo) {
		if (nodo == nullptr) {
			return;
		}
		_enPosOrden(nodo->izq);
		_enPosOrden(nodo->der);
		cout<<nodo->elemento<<" ";
	}
	bool _vacio() {
		return raiz == nullptr;
	}
  int _cantidad(Nodo<T>* nodo) {
		if (nodo==nullptr)
			return 0;
		else
		{
			int ci, cd;
			ci = _cantidad(nodo->izq);
			cd = _cantidad(nodo->der);
			return 1 + ci + cd;
		}
	}
  T _obtener(Nodo<T>* nodo, T ele){
      if(nodo==nullptr){
          return 0;
      }
      else{
					int res=comparar(nodo->elemento, ele);
					if(res==0){
							return nodo->elemento;
					}
					else if(res<0){
							return _obtener(nodo->der, ele);
					}else{
							return _obtener(nodo->izq, ele);
					}          
      }      
  }
	bool _eliminar(Nodo<T>* nodo, T ele){
			if(nodo==nullptr){
					return false;
			}
			else{
					int res=comparar(nodo->elemento, ele);
					if(res<0){
							return _eliminar(nodo->der,ele);
					}else if(res>0){
							return _eliminar(nodo->izq,ele);
					}else{
							if(nodo->der==nullptr && nodo->izq==nullptr){
									nodo = nullptr;
                  delete nodo;      
                  return true;
							}else if(nodo->izq==nullptr){
									nodo=nodo->der; 
									return true;									
							}else if(nodo->der==nullptr){
									nodo=nodo->izq; 
									return true;
							}else{
									Nodo<T>* aux= nodo->der;
									while(aux->izq!=nullptr){
											aux=aux->izq;
									}
									nodo->elemento=aux->elemento;
									return _eliminar(nodo->der,aux->elemento);
							}
					}
			}
	}
	bool _buscar(Nodo<T>* nodo, T e){
			if(nodo==nullptr){
					return false;
			}else{
					int r = comparar(nodo->elemento,e);
					if(r==0){
							return true;
					}else if(r<0){
							return _buscar(nodo->der,e);
					}else{
							return _buscar(nodo->izq,e);
					}
			}
	}
	short _altura(Nodo<T>* nodo) {
			short iz=0; short de=0;
			if (nodo == nullptr){
					return 0;
			}else{           
            iz = 1 + _altura(nodo->izq);
            de = 1 + _altura(nodo->der);
            return max(iz,de);
        }
    }
public:
	ArbolB(function<int(T,T)> comparar) {	
		raiz = nullptr;
		this->comparar=comparar;
	}
	bool insertar(T e) {
		return _insertar(raiz, e);
	}
	void enOrden() {
		_enOrden(raiz);
  }
	void enPreOrden() {
		_enPreOrden(raiz);
  }
	void enPosOrden() {
		_enPosOrden(raiz);
  }
  int cantidad() {
		return _cantidad(raiz);
	}
	short altura() {
		return _altura(raiz);
	}
  T obtener(T e){
			return _obtener(raiz,e);
	}
	bool eliminar(T e){
			return _eliminar(raiz,e);
	}
	bool buscar(T e){
			return _buscar(raiz,e);
	} 
};
```
